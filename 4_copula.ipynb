# Gaussian Copula

A **copula** is a function that links multivariate distribution functions to their one-dimensional marginal distribution functions. The Gaussian copula is one of the most common types of copulas and is based on the multivariate normal distribution.

### Definition

The Gaussian copula is defined as follows:

$ \[C(u_1, u_2, ..., u_n) = \Phi_{\Sigma}(\Phi^{-1}(u_1), \Phi^{-1}(u_2), ..., \Phi^{-1}(u_n))\] $

where $\( \Phi \)$ is the standard normal cumulative distribution function (CDF), and $\( \Phi_{\Sigma} \)$ is the joint CDF of a multivariate normal distribution with zero mean and covariance matrix $\( \Sigma \) $.

### Visualization

In this notebook, we will:
- Define a Gaussian copula
- Generate random samples
- Visualize the copula using a scatter plot

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Parameters
mean = [0, 0]  # Mean of the multivariate normal
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix
size = 1000  # Number of samples
```

```python
# Generate samples from a multivariate normal distribution
data = np.random.multivariate_normal(mean, cov, size)
```

```python
# Transform data using the CDF of the normal distribution
u = stats.norm.cdf(data)
```

```python
# Plot
plt.figure(figsize=(5, 4))
plt.scatter(u[:, 0], u[:, 1], alpha=0.5)
plt.title('Scatter Plot of Gaussian Copula')
plt.xlabel('U1')
plt.ylabel('U2')
plt.grid(True)
plt.show()
```

This structure above is based on the multivariate normal distribution, reflecting the co-movement characteristics modeled by the Gaussian copula
